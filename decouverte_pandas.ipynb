## 1. Un DataFrame = un tableau

Un **DataFrame** pandas, c'est une table : **1 ligne = 1 observation**
(ici un ticket), **1 colonne = 1 variable**. On en fabrique un à la main
pour l'exemple — en vrai il viendra d'un fichier (voir cellule suivante).

```python
import pandas as pd

tickets = pd.DataFrame({
    "date":      ["2026-06-01", "2026-06-01", "2026-06-02", "2026-06-02",
                  "2026-06-03", "2026-06-03", "2026-06-04", "2026-06-04",
                  "2026-06-05", "2026-06-05", "2026-06-05", "2026-06-06"],
    "client":    ["Pyrenex", "Eckmühl", "Pyrenex", "Athéna RH", "Eckmühl", "Pyrenex",
                  "Athéna RH", "Eckmühl", "Pyrenex", "Athéna RH", "Eckmühl", "Pyrenex"],
    "canal":     ["mail", "téléphone", "mail", "chat", "mail", "téléphone",
                  "chat", "mail", "téléphone", "mail", "chat", "mail"],
    "priorite":  ["haute", "basse", "moyenne", "haute", "basse", "moyenne",
                  "haute", "moyenne", "basse", "haute", "moyenne", "basse"],
    "resolu":    [True, True, False, True, True, False,
                  True, True, False, True, True, True],
    "duree_min": [42, 15, 60, 25, 10, 75, 30, 20, 55, 28, 18, 12],
})
tickets
```

## 2. En vrai, les données viennent d'un fichier : `read_csv`

Le format le plus courant est le **CSV** (valeurs séparées par des virgules).
On sauvegarde notre table (`to_csv`), puis on la **relit** (`read_csv`) — c'est
exactement le geste que tu feras lundi sur le dataset fourni.

```python
# Écrire puis relire — le 'aller-retour' CSV
tickets.to_csv("tickets_fastia.csv", index=False)   # index=False : pas de colonne d'index parasite

df = pd.read_csv("tickets_fastia.csv")
print("Chargé :", df.shape[0], "lignes ×", df.shape[1], "colonnes")
```

## 3. Regarder : `head` et `shape`

Premier réflexe sur un dataset inconnu : **en voir un bout** et connaître sa taille.

```python
df.head()      # les 5 premières lignes (df.head(3) pour 3)
```

## 4. Comprendre les colonnes : `info` et les types

`info()` liste les colonnes, leur **type** (`dtype`) et les valeurs manquantes.
Un piège classique en M1 : une colonne qu'on croit numérique mais que pandas a
lue comme du texte (`object`).

```python
df.info()
```

## 5. Statistiques rapides : `describe`

`describe()` résume les colonnes **numériques** (moyenne, min, max, quartiles).
Ici seule `duree_min` est numérique.

```python
df.describe()
```

## 6. Compter par catégorie : `value_counts`

**Le geste EDA le plus utile.** Combien de tickets par canal ? par priorité ?
C'est ce qui te permettra lundi de repérer une **classe minoritaire**
(le fameux *déséquilibre* dont parle le carnet M1).

```python
df["canal"].value_counts()
```

## 7. Agréger : `groupby`

Regrouper les lignes par une colonne, puis calculer sur chaque groupe.
Question métier : **la durée moyenne de résolution dépend-elle de la priorité ?**

```python
df.groupby("priorite")["duree_min"].mean().round(1)
```

## 8. Une visualisation en une ligne

pandas sait tracer directement (via matplotlib). Un graphe vaut souvent
mieux qu'un tableau pour *voir* une répartition.

```python
# nécessite matplotlib (déjà dans l'env M1)
df["canal"].value_counts().plot(kind="bar", title="Tickets par canal");
```

## 🟢 Ce que tu retiens

| Geste | À quoi ça sert |
|---|---|
| `pd.read_csv(...)` | charger un tableau depuis un fichier |
| `df.head()` / `df.shape` | jeter un œil, connaître la taille |
| `df.info()` | colonnes, types, valeurs manquantes |
| `df.describe()` | stats rapides du numérique |
| `df["col"].value_counts()` | compter par catégorie (repérer un déséquilibre) |
| `df.groupby("col")["x"].mean()` | agréger par groupe |
