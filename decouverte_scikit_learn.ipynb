L'objectif ici n'est **pas** d'apprendre à entraîner un modèle correctement
— *ça, c'est le geste de M1-B1, lundi*. On veut juste que tu voies tourner
**une fois** les 3 verbes de scikit-learn, pour arriver lundi sans choc de syntaxe :

| Verbe | Ce qu'il fait |
|---|---|
| `.fit(X, y)` | **entraîner** — le modèle apprend sur des données dont on connaît la réponse |
| `.predict(X)` | **prédire** — le modèle répond sur de nouvelles lignes |
| `.score(X, y)` | **noter** — un chiffre qui résume la qualité |

Tous les modèles scikit-learn partagent cette même grammaire. Une fois ces
3 verbes intégrés, changer de modèle ne change que **deux lignes**.

```python
# Versions testées : scikit-learn 1.8, pandas 3.0 (la version exacte importe peu)
from sklearn.datasets import load_wine

# Un petit dataset EMBARQUÉ dans scikit-learn — aucun téléchargement.
# 178 vins, 13 mesures chimiques, 3 cépages à reconnaître.
data = load_wine(as_frame=True)
X = data.data      # les FEATURES (variables d'entrée) — ce qu'on donne au modèle
y = data.target    # la CIBLE (y) — le cépage à prédire : 0, 1 ou 2

print("Forme de X :", X.shape, " -> 178 lignes x 13 features")
print("Cibles possibles (y) :", sorted(y.unique()))

# Un dataset tabulaire = un tableau : 1 ligne = 1 observation, 1 colonne = 1 feature
X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

# random_state=42 -> résultat REPRODUCTIBLE (même code = même modèle)
model = RandomForestClassifier(random_state=42)

# .fit() = ENTRAÎNER : le modèle ajuste ses paramètres internes sur (X, y)
model.fit(X, y)
```

```python
import pandas as pd

# .predict() = PRÉDIRE. On regarde 6 vins répartis sur les 3 cépages.
lignes = [0, 40, 70, 110, 140, 175]
comparaison = pd.DataFrame(
    {"réel (y)": y.iloc[lignes].values,
     "prédit par le modèle": model.predict(X.iloc[lignes])},
    index=lignes,
)
comparaison
```

```python
# .score() = une NOTE globale (ici : proportion de bonnes réponses = accuracy)
note = model.score(X, y)
print(f"Score sur les données d'entraînement : {note:.2%}")
```
